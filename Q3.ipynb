```python
import numpy as np
import math
import matplotlib.pyplot as plt
```

```python
G=6.673*(10**-11)
M=2*(10**30)
m1=6*(10**24)
m2=3*(10**24)
u1=G*(M+m1)
u2=G*(M+m2)
w2 = 2*(10**-7) ##earth's avg. angular velocity
w1 = w2
a1 = 1.5*(10**11)
a2 = 1.2*(10**11)
e1 = 0.3
e2 = 0.4
```

```python
h1 = w1*a1*a1
h2 = w2*a2*a2
output_r1 = np.zeros(360)
output_r2 = np.zeros(360)
output_theta = np.zeros(360)
```

```python
output_r1[0] = (h1*h1)/(u1*(1+e1*math.cos(output_theta[0])))
output_r2[0] = (h2*h2)/(u2*(1+e2*math.cos(output_theta[0])))
```

```python
for x in range(1,360):
    output_theta[x] = output_theta[x-1] + 1*math.pi/180; 
    output_r1[x] = (h1*h1)/(u1*(1+e1*math.cos(output_theta[x])))
    output_r2[x] = (h2*h2)/(u2*(1+e2*math.cos(output_theta[x])))

print(output_theta)
print(output_r1)
```

```python
# plotting
plt.title("Variation of r wrt theta")
plt.xlabel("theta")
plt.ylabel("radial distance")
plt.plot(output_theta,output_r1, color ="red")
plt.plot(output_theta,output_r2, color ="blue")
plt.show()
```

```python
#relative positions

output_r12 = np.subtract(output_r1,output_r2)
output_r21 = np.subtract(output_r2,output_r1)
```

```python
# plotting
plt.title("Variation of m1 wrt m2(red) and m2 wrt m1(blue)")
plt.xlabel("theta")
plt.ylabel("radial distance")
plt.plot(output_theta,output_r12, color ="red")
plt.plot(output_theta,output_r21, color ="blue")
plt.show()
```
