```python
import numpy as np
import math
import matplotlib.pyplot as plt
```

```python
G=6.673*(10**-11)
M=2*(10**30)
n = 2
w2 = 2*(10**-7) ##earth's avg. angular velocity
T = 0
if (n==1):
    T = 0
else:
    T = 5184000 ##(in seconds)Random value = 60 days
w1 = w2*n
theta_zero = (n-1)*w1*T
r1 = ((G*M)/(w1*w1))**(1./3.)
r2 = ((G*M)/(w2*w2))**(1./3.)
T1 = (2*math.pi)/w1 ##Time period of m1
T2 = (2*math.pi)/w2 ##Time period of m2
```

```python
resolution = 1000;
time = np.zeros(resolution)
output_r1x = np.zeros(resolution)
output_r1y = np.zeros(resolution)
output_r2x = np.zeros(resolution)
output_r2y = np.zeros(resolution)
time_step = (T1+T2)/resolution/2
```

```python
##Computing for 1 year of m2
time_step = T2/resolution
for x in range(1,resolution):
    time[x] = time[x-1]+time_step

for x in range(0,resolution):
    output_r1x[x] = r1*math.cos(w1*time[x])
    output_r1y[x] = r1*math.sin(w1*time[x])
    
for x in range(0,resolution):
    output_r2x[x] = r2*math.cos(w2*time[x]+theta_zero)
    output_r2y[x] = r2*math.sin(w2*time[x]+theta_zero)
    
output_r12x = np.zeros(resolution)
output_r12y = np.zeros(resolution)

output_r12x = np.subtract(output_r1x,output_r2x)
output_r12y = np.subtract(output_r1y,output_r2y)

# plotting
plt.title("Variation of m1 wrt m2")
plt.xlabel("time")
plt.ylabel("position in x(red) and y(blue)")
plt.plot(time,output_r12x, color ="red")
plt.plot(time,output_r12y, color ="blue")
plt.show()
```
